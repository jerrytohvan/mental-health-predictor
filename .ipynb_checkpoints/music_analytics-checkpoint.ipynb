# Spotify Music Analytics on Mental Health

```python
# Import libraries
import numpy as np
import pandas as pd
from IPython.display import display, HTML

from textblob import TextBlob
#https://medium.com/@rahulvaish/textblob-and-sentiment-analysis-python-a687e9fabe96 , use this to perform sentiment polarity to derive sentiment classification on lyrics
```

Text.

```python
# Read the census data into a DataFrame named census_df
spotify_survey_df = pd.read_csv('./data/master.csv')
spotify_survey_df.head()
```

## Feature Encoding

sklearn only accepts numerical values as input. So we use `LabelEncoder` to encode all features as into numeric values.

Documentation: https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.LabelEncoder.html

```python
from sklearn import preprocessing
```

```python
for column in spotify_survey_df.columns:
    if spotify_survey_df[column].dtype == type(object):
        le = preprocessing.LabelEncoder()
        spotify_survey_df[column] = le.fit_transform(spotify_survey_df[column])
        
#removing NaN values on rows. Alternative perform mean on column values but actually it doesnt realy value add much, so its better to remove.
spotify_survey_df = spotify_survey_df.dropna()

print("Dataframe:")
display(HTML(spotify_survey_df.to_html()))
```

## Train-Test Split

Assign the features to `X` and the target to `y`.

```python
# features
X = spotify_survey_df.iloc[: , [14,15,16,17,18,19,20,21]]

# classifier
y = spotify_survey_df['health_categorical']


print("Dataframe:")
display(HTML(X.to_html()))


```

Create a training set containing a random sample of 80% of the observations, and a test set with the remaining 20%. Set the `random_state` to `424`.

```python
from sklearn.model_selection import train_test_split
```

```python
# split into training and test set
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size =0.2,random_state =424)
```

## Performing EDA on Current Data



## Classification using Decision Tree

Decision Tree results are weird. Not appropriate model

Documentation: https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html#sklearn.tree.DecisionTreeClassifier.fit

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# initialise Decision Tree
dtree = DecisionTreeClassifier(criterion='entropy', max_depth=2, random_state=424)

# train model
dtree.fit(X_train, y_train)
```

Plot the decision tree using `plot_tree`. How many leaf nodes does the tree have? *Answer: 4 leaf nodes*

Documentation: https://scikit-learn.org/stable/modules/generated/sklearn.tree.plot_tree.html

```python
from sklearn.tree import plot_tree
```

```python
plot_tree(dtree.fit(X_test, y_test))
```

Get predictions of the test set using `predict`.

```python
# get predictions 
predictions = dtree.predict(X_test)

# compare results
pd.DataFrame({'Actual': y_test, 'Predicted': predictions}) \
    .sample(10)
```

Retrieve the confusion matrix using `confusion_matrix` and print the results obtained.

Documentation:  https://scikit-learn.org/stable/modules/generated/sklearn.metrics.confusion_matrix.html

Get the true negatives (TN), false negatives (FN), true positives (TP) and false positives (FP) from the confusion matrix. What is the accuracy in the test set? What is the f1-score (f measure) of the test set?

```python
from sklearn.metrics import confusion_matrix
```

```python
# Get the confusion matrix
conf_matrix = confusion_matrix(y_test, predictions)
conf_matrix
```

```python
# Retrieve TN, FN, TP and FP
tn = conf_matrix[0][0]
fn = conf_matrix[1][0]
tp = conf_matrix[1][1]
fp = conf_matrix[0][1]
```

Compute accuracy and f1 score with the TN, FN, TP and FP values. (Refer to week 5 slides). 
Another ref: https://towardsdatascience.com/accuracy-recall-precision-f-score-specificity-which-to-optimize-on-867d3f11124

```python
#Accuracy = (TP+TN)/(TP+FP+FN+TN)
accuracy = (tp+tn)/(tp+fp+fn+tn)
```

```python
#Precision = TP/(TP+FP)
precision = tp/(tp+fp)

#Recall = TP/(TP+FN)
recall = tp/(tp+fn)

#F1 Score = 2*(Recall * Precision) / (Recall + Precision)
f1_score = 2*(recall*precision)/(recall+precision)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F-Score:", f1_score)
```

## Classification using Random Forest

Random classification results to a weird perfect result.

Documentation: https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html

```python
# Your code here

#import library
from sklearn.ensemble import RandomForestClassifier

# initialise classifier, default n_estimator: 100
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=424)

# train model
rf_classifier.fit(X_train, y_train)
```

```python
# get predictions 
rf_predictions = rf_classifier.predict(X_test)

# compare results
pd.DataFrame({'Actual': y_test, 'Predicted': rf_predictions}) \
    .sample(10)
```

```python
# Get the confusion matrix
rf_conf_matrix = confusion_matrix(y_test, rf_predictions)
print(rf_conf_matrix)

# Retrieve TN, FN, TP and FP
rf_tn = rf_conf_matrix[0][0]
rf_fn = rf_conf_matrix[1][0]
rf_tp = rf_conf_matrix[1][1]
rf_fp = rf_conf_matrix[0][1]


#Accuracy = (TP+TN)/(TP+FP+FN+TN)
rf_accuracy = (rf_tp+rf_tn)/(rf_tp+rf_fp+rf_fn+rf_tn)

#Precision = TP/(TP+FP)
rf_precision = rf_tp/(rf_tp+rf_fp)

#Recall = TP/(TP+FN)
rf_recall = rf_tp/(rf_tp+rf_fn)

#F1 Score = 2*(Recall * Precision) / (Recall + Precision)
rf_f1_score = 2*(rf_recall*rf_precision)/(rf_recall+rf_precision)


print("\n======= USING MANUAL FORMULA ========")
print("Accuracy:", rf_accuracy)
print("Precision:", rf_precision)
print("Recall:", rf_recall)
print("F-Score:", rf_f1_score)

from sklearn import metrics

print("\n======= USING SKLEARN METRICS LIBRARY ========")
print("Accuracy:",metrics.accuracy_score(y_test, rf_predictions))
print("Precision:",metrics.precision_score(y_test, rf_predictions))
print("Recall:",metrics.recall_score(y_test, rf_predictions))
print("F-Score:",metrics.f1_score(y_test, rf_predictions))
```

```python
# Get the confusion matrix
nn_conf_matrix = confusion_matrix(y_test, nn_predictions)
print(rf_conf_matrix)

# Retrieve TN, FN, TP and FP
nn_tn = nn_conf_matrix[0][0]
nn_fn = nn_conf_matrix[1][0]
nn_tp = nn_conf_matrix[1][1]
nn_fp = nn_conf_matrix[0][1]


#Accuracy = (TP+TN)/(TP+FP+FN+TN)
nn_accuracy = (nn_tp+nn_tn)/(nn_tp+nn_fp+nn_fn+nn_tn)

#Precision = TP/(TP+FP)
nn_precision = nn_tp/(nn_tp + nn_fp)

#Recall = TP/(TP+FN)
nn_recall = nn_tp/(nn_tp+ nn_fn)

#F1 Score = 2*(Recall * Precision) / (Recall + Precision)
nn_f1_score = 2*(nn_recall*nn_precision)/(nn_recall+nn_precision)


print("\n======= USING MANUAL FORMULA ========")
print("Accuracy:", nn_accuracy)
print("Precision:", nn_precision)
print("Recall:", nn_recall)
print("F-Score:", nn_f1_score)

from sklearn import metrics

print("\n======= USING SKLEARN METRICS LIBRARY ========")
print("Accuracy:",metrics.accuracy_score(y_test, nn_predictions))
print("Precision:",metrics.precision_score(y_test, nn_predictions))
print("Recall:",metrics.recall_score(y_test, nn_predictions))
print("F-Score:",metrics.f1_score(y_test, nn_predictions))
```
